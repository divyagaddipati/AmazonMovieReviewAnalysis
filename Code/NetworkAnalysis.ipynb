# Imports

```python
import networkx as nx
import operator
import matplotlib.pyplot as plt
import networkx.algorithms.community as nxcom
```

```python
G = nx.read_edgelist('NetworkAnalysis_src_target_reviews_Movies_and_TV_5.txt')
```

```python
plt.figure(figsize=(5,5))
nx.draw_networkx(G)
```

### Basic Analysis

```python
num_nodes, num_edges = G.order(), G.size()
avg_deg = num_edges / num_nodes
conn_comps = nx.number_connected_components(G)

# number of nodes and edges

print("Number of nodes : ", str(num_nodes))

print("Number of edges : ", str(num_edges))

print("Average degree: ", avg_deg)

print("connected_components: ", conn_comps)

print("Density", nx.density(G))
```

```python
# diameter of graph
nx.diameter(G)
```

```python
# Average Shortest Path
nx.average_shortest_path_length(G)
```

```python
# Number of communities in whole graph
communities = sorted(nxcom.greedy_modularity_communities(G), key=len, reverse=True)
print(len(communities))
```

```python
nxcom.modularity(G, communities)
```

### Centrality Analysis



```python
# degree centrality (10 nodes that are most 'important' based on each centrality measure)
dc = nx.degree_centrality(G)
sorted_dc = sorted(dc.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_dc[0:9])
```

```python
# closeness centrality (10 nodes that are most 'important' based on each centrality measure)
cc = nx.closeness_centrality(G)
sorted_cc = sorted(cc.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_cc[0:9])
```

```python
# betweenness centrality (10 nodes that are most 'important' based on each centrality measure)
cc = nx.betweenness_centrality(G)
sorted_bc = sorted(cc.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_bc[0:9])
```

### Clustering Coefficient Analysis

```python
print(nx.average_clustering(G))
```

```python
# Plot of frequency of clustering coefficients in graph
plt.hist(nx.clustering(G).values())
plt.title('Clustering Vs. Frequency')
plt.ylabel('Frequency')
plt.xlabel('Clustering')
plt.show()
```
