```python
### Source : http://jmcauley.ucsd.edu/data/amazon/index_2014.html

import pandas as pd
import gzip
import json

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
      yield json.loads(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
      df[i] = d
      i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df = getDF('../reviews_Movies_and_TV_5.json.gz')
```

```python
# df reviewers with >= 100 reviews
# 1. To reduce run time 
# 2. Get the reviewers with a significant amount of reviews

counts = df['reviewerID'].value_counts()
df = df[df['reviewerID'].isin(counts[counts >= 100].index)]
```

```python
df_reviewer_movies = df.filter(['reviewerID', 'asin'], axis=1)
df_reviewer_movies.rename(columns = {'reviewerID':'source', 'asin':'target'}, inplace = True)


df_reviewer_movies = df_reviewer_movies.set_index('source')
```
